```python
import heapq

def ida_star(graph, heuristic, start, goal, path):
    threshold = heuristic[start]  # Initial threshold
    while True:
        steps, new_threshold = search(graph, heuristic, start, goal, path, threshold)
        if steps is not None:  # If goal found within threshold, return the steps
            return steps
        elif new_threshold == float('inf'):  # If threshold is infinity, no solution
            return None
        threshold = new_threshold  # Increase threshold for next iteration

def search(graph, heuristic, start, goal, path, threshold):
    frontier = [(heuristic[start], 0, start)]  # Initialize frontier with start node
    explored = set()  # Initialize set to keep track of explored nodes
    checking_thresh=[]

    while frontier:
        _, cost, node = heapq.heappop(frontier)  # Pop node with the lowest f-value
        if node == goal:
            print('Goal is reached')# If goal reached within threshold, return the path and cost
            return path, cost
        if cost + heuristic[node] <= threshold:  # If node cost + heuristic <= threshold, explore neighbors
            explored.add(node)
            for neighbor, weight in graph[node].items():
                if neighbor not in explored:
                    new_cost = cost + weight
                    checking_thresh.append(new_cost+ heuristic[neighbor])
                    heapq.heappush(frontier, (new_cost + heuristic[neighbor], new_cost, neighbor))
                    path.append(neighbor)  # Add neighbor to path
    return None, min(checking_thresh)  # Return None and the new threshold if frontier is empty

# Take user input for graph
graph = {}
heuristic = {}

num_edges = int(input("Enter the number of edges: "))
print("Enter edges and their costs (node1 node2 cost(g(n))):")
for _ in range(num_edges):
    node1, node2, cost = input().split()
    cost = int(cost)
    if node1 not in graph:
        graph[node1] = {}
    if node2 not in graph:
        graph[node2] = {}
    graph[node1][node2] = cost
    graph[node2][node1] = cost
    

# Take user input for heuristic value
num_nodes = int(input("Enter the number of nodes: "))
print("Enter heuristic values for each node (node heuristic_value (h(n))):")
for _ in range(num_nodes):
    node, heuristic_value = input().split()
    heuristic[node] = int(heuristic_value)

# Take user input for source node and goal node
start_node = input("Enter the source node: ")
goal_node = input("Enter the goal node: ")

path = []
ida_star(graph, heuristic, start_node, goal_node, path)
print("Path:", path)

```
