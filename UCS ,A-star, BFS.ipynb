```python
import heapq

def uniform_cost_search(graph, start, goal, path):
    frontier = [(0, start)]
    explored = set()
    steps = []

    while frontier:
        cost, node = heapq.heappop(frontier)
        steps.append((node, cost))
        if node in goal:
            return steps
        if node not in explored:
            explored.add(node)
            for neighbor, weight in graph[node].items():
                if neighbor not in explored:
                    heapq.heappush(frontier, (cost + weight, neighbor))
                    path.append(neighbor)

    return steps

def a_star(graph, heuristic, start, goal, path):
    frontier = [(heuristic[start], 0, start)]
    explored = set()
    steps = []

    while frontier:
        _, cost, node = heapq.heappop(frontier)
        steps.append((node, cost))
        if node == goal:
            return steps
        if node not in explored:
            explored.add(node)
            for neighbor, weight in graph[node].items():
                if neighbor not in explored:
                    heapq.heappush(frontier, (heuristic[neighbor] + cost + weight, cost + weight, neighbor))
                    path.append(neighbor)

    return steps


def best_first_search(graph, heuristic, start, goal, path):
    frontier = [(heuristic[start], start)]
    explored = set()
    steps = []

    while frontier:
        _, node = heapq.heappop(frontier)
        steps.append((node, heuristic[node]))
        if node == goal:
            return steps
        if node not in explored:
            explored.add(node)
            for neighbor, _ in graph[node].items():
                if neighbor not in explored:
                    heapq.heappush(frontier, (heuristic[neighbor], neighbor))
                    path.append(neighbor)

    return steps

# Take user input for graph
graph = {}
heuristic = {}

num_edges = int(input("Enter the number of edges: "))
print("Enter edges and their costs (node1 node2 cost(g(n))):")
for _ in range(num_edges):
    node1, node2, cost = input().split()
    cost = int(cost)
    if node1 not in graph:
        graph[node1] = {}
    if node2 not in graph:
        graph[node2] = {}
    graph[node1][node2] = cost
    graph[node2][node1] = cost

    
# Take user input for heuristic value   
num_nodes = int(input("Enter the number of nodes: "))
print("Enter heuristic values for each node (node heuristic_value (h(n))):")
for _ in range(num_nodes):
    node, heuristic_value = input().split()
    heuristic[node] = int(heuristic_value)

goal_node=[]
    
# Take user input for source node and goal node
start_node = input("Enter the source node: ")

n=int(input("Enter the no. goal nodes :"))
for _ in range (n):
    a=input("Enter the goal node: ")
    goal_node.append(a)

path=[]
# Perform search algorithms and show intermediate steps
print("\nUniform Cost Search:")
uc_steps = uniform_cost_search(graph, start_node, goal_node, path)
for step in uc_steps:
    print(step)

print("\nA* algorithm:")
a_star_steps = a_star(graph, heuristic, start_node, goal_node, path)
for step in a_star_steps:
    print(step)

print("\nBest First Search:")
best_first_steps = best_first_search(graph, heuristic, start_node, goal_node, path)
for step in best_first_steps:
    print(step)
```
